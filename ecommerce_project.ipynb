```python
#RFM ANALYSIS
import datetime as dt

latest_date = df['order_date'].max()

rfm = df.groupby('customer_id').agg({
    'order_date': lambda x: (latest_date - x.max()).days,
    'order_id': 'count',
    'sales': 'sum'
})
```

```python
rfm.columns = ['recency', 'frequency', 'monetary']
rfm.head()
```

```python
#RFM SCORING
```

```python
rfm['r_score'] = pd.cut(rfm['recency'], bins=4, labels=[4,3,2,1])
rfm['f_score'] = pd.cut(rfm['frequency'], bins=4, labels=[1,2,3,4])
rfm['m_score'] = pd.cut(rfm['monetary'], bins=4, labels=[1,2,3,4])
```

```python
rfm.head()
```

```python
#CUSTOMER SEGMENTS
rfm['segment'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str)
rfm.head()
```
